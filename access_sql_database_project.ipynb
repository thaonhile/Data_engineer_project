### Store the dataset in a Table

##### In many cases the dataset to be analyzed is available as a .CSV (comma separated values) file, perhaps on the internet.  
##### To analyze the data using SQL, it first needs to be stored in the database.

##### We will first read the csv files  from the given url  into pandas dataframes

##### Next we will be using the  df.to_sql() function to convert each csv file  to a table in sqlite  with the csv data loaded in it.


```python
import pandas as pd
df= pd.read_csv('https://data.cityofchicago.org/resource/9xs2-f89t.csv')
df.to_sql('CHICAGO_PUBLIC_SCHOOLS_DATA',con, if_exists ='replace', index = False, method='multi')
```

## Problems

### Problem 1

##### How many Elementary Schools are in the dataset?


```python
%sql select distinct count(name_of_school)  from 'CHICAGO_PUBLIC_SCHOOLS_DATA' where elementary_or_high_school = 'ES'
```

### Problem 2

##### What is the highest Safety Score?


```python

%sql select max(safety_score)  from 'CHICAGO_PUBLIC_SCHOOLS_DATA'
```

### Problem 3

##### Which schools have highest Safety Score?


```python
%sql select name_of_school , average_student_attendance from 'CHICAGO_PUBLIC_SCHOOLS_DATA' where safety_score = (select max(safety_score)  from 'CHICAGO_PUBLIC_SCHOOLS_DATA')
```

### Problem 4

##### What are the top 10 schools with the highest "Average Student Attendance"?


```python
%sql select name_of_school,average_student_attendance from 'CHICAGO_PUBLIC_SCHOOLS_DATA' order by average_student_attendance  desc  limit 10
```

### Problem 5

##### Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance


```python
%sql select name_of_school , average_student_attendance from 'CHICAGO_PUBLIC_SCHOOLS_DATA' order by average_student_attendance limit 5
```

### Problem 6

##### Now remove the '%' sign from the above result set for Average Student Attendance column


```python
%sql select name_of_school , replace(average_student_attendance,'%','') from 'CHICAGO_PUBLIC_SCHOOLS_DATA' 
```

### Problem 7

##### Which Schools have Average Student Attendance lower than 70%?


```python
%sql select name_of_school,average_student_attendance from 'CHICAGO_PUBLIC_SCHOOLS_DATA' where average_student_attendance <70
```

### Problem 8

##### Get the total College Enrollment for each Community Area


```python
%sql select community_area_name	, sum(college_enrollment_rate) as total_enrollment from 'CHICAGO_PUBLIC_SCHOOLS_DATA' group by community_area_name	
```

### Problem 9

##### Get the 5 Community Areas with the least total College Enrollment  sorted in ascending order


```python
%sql select community_area_name	, sum(college_enrollment_rate) as total_enrollment from 'CHICAGO_PUBLIC_SCHOOLS_DATA' group by community_area_name	 order by sum(college_enrollment_rate) limit 5
```

### Problem 10

##### List 5 schools with lowest safety score.


```python
%sql select name_of_school , safety_score from 'CHICAGO_PUBLIC_SCHOOLS_DATA' where safety_score != 'None' order by safety_score limit 5 
```

### Problem 11

##### Get the community area of the school which has College Enrollment of 42.9


```python
%sql select community_area_name	, sum(college_enrollment_rate) as total_enrollment from 'CHICAGO_PUBLIC_SCHOOLS_DATA'  group by community_area_name having sum(college_enrollment_rate) = 42.9
```
